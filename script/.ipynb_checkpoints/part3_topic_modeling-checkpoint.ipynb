```python
df = pd.read_csv('tweets.csv',index_col='Unnamed: 0')
df.dropna(subset=['text'], inplace=True)
df
```

```python
from transformers import pipeline
from tqdm.auto import tqdm

sentiment_analysis = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
def analyze_sentiment(text):
    try:
        result = sentiment_analysis(text)[0]
        score = result['score']
        if result['label'] == 'POSITIVE' and score > 0.55:  # ÈòàÂÄºÂèØ‰ª•Ë∞ÉÊï¥
            return 1
        elif result['label'] == 'NEGATIVE' and score > 0.55:  # ÈòàÂÄºÂèØ‰ª•Ë∞ÉÊï¥
            return -1
        else:
            return 0
    except Exception as e:
        print(f"Error in sentiment analysis: {e}")
        return 'error'
```

```python
tqdm.pandas()
df['sentiment'] = df['text'].progress_apply(analyze_sentiment)
df.to_csv("1116_save.csv")
```

```python
# ÈíàÂØπÊñáÊú¨ËøõË°åËÅöÁ±ª
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Â∞ÜÊñáÊú¨Êï∞ÊçÆËΩ¨Êç¢‰∏∫ TF-IDF ÁâπÂæÅ
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(df['text'])
```

```python
# ‰ΩøÁî®ËÇòÈÉ®Ê≥ïÂàôÊâæÂà∞ÊúÄ‰Ω≥ÁöÑÁ∞áÊï∞Èáè
wcss = []
silhouette_scores = []
K = range(2, 10)
for k in K:
    print(k)
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X, kmeans.labels_))
```

```python
# ÁªòÂà∂ËÇòÈÉ®Ê≥ïÂàôÂõæ
plt.figure(figsize=(12, 6))
plt.plot(K, wcss, 'bo-')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.title('The Elbow Method showing the optimal number of clusters')
plt.show()
```

```python
# ÁªòÂà∂ËΩÆÂªìÂàÜÊï∞Âõæ
plt.figure(figsize=(12, 6))
plt.plot(K, silhouette_scores, 'ro-')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for each number of clusters')
plt.show()
```

```python
import numpy as np

# ÊâæÂà∞ËÇòÈÉ®ÁÇπ
optimal_k_elbow = np.argmin(np.diff(np.diff(wcss))) + 2  # ‰ΩøÁî®‰∫åÈò∂ÂØºÊï∞ÁöÑÊîπÂèò

# ÊâæÂà∞ËΩÆÂªìÂàÜÊï∞ÊúÄÂ§ßÁöÑÁÇπ
optimal_k_silhouette = silhouette_scores.index(max(silhouette_scores)) + 2

# ËæìÂá∫ÊúÄ‰Ω≥ËÅöÁ±ªÊï∞Èáè
optimal_k = max(optimal_k_elbow, optimal_k_silhouette)

print(f"Optimal number of clusters determined by Elbow Method: {optimal_k_elbow}")
print(f"Optimal number of clusters determined by Silhouette Score: {optimal_k_silhouette}")
print(f"Chosen optimal number of clusters: {optimal_k}")
```

```python
# ËøõË°åËÅöÁ±ª
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans.fit(X)
df['cluster'] = kmeans.labels_
```

```python
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import TfidfVectorizer
import seaborn as sns

for cl in range(optimal_k):
    print("cluster"+str(cl))
    # ÂÅáËÆæ df ÊòØÊÇ®ÁöÑ DataFrameÔºå‰∏îÂåÖÂê´ 'text' Âíå 'cluster' Âàó
    # ÂÅáËÆæÊàë‰ª¨ÂÖ≥Ê≥®ÁöÑÊòØÁ¨¨‰∏Ä‰∏™ËÅöÁ±ª
    cluster_docs = df[df['cluster'] == cl]['text']
    
    # Â∞ÜÊñáÊú¨ËΩ¨Êç¢‰∏∫ÁâπÂæÅ
    vectorizer = TfidfVectorizer(stop_words='english')
    X_cluster = vectorizer.fit_transform(cluster_docs)
    
    # # ËØïÈ™å‰∏çÂêåÁöÑ‰∏ªÈ¢òÊï∞Èáè‰ª•Á°ÆÂÆöÊúÄ‰Ω≥‰∏ªÈ¢òÊï∞Ôºå‰ΩÜ‰∏∫‰∫ÜÊï¥ÈΩêËµ∑ËßÅÔºåÁªü‰∏ÄÊòØ5
    # best_score = -np.inf
    # best_k = 0
    # for k in range(2, 10):  # ÂèØ‰ª•Ê†πÊçÆÈúÄË¶ÅË∞ÉÊï¥ËåÉÂõ¥
    #     print(k)
    #     lda = LatentDirichletAllocation(n_components=k, random_state=42)
    #     lda.fit(X_cluster)
    #     score = lda.score(X_cluster)
    #     if score > best_score:
    #         best_score = score
    #         best_k = k
    
    best_k = 5
    
    # ‰ΩøÁî®ÊúÄ‰Ω≥‰∏ªÈ¢òÊï∞ÈáèÈáçÊñ∞Â∫îÁî® LDA
    lda_best = LatentDirichletAllocation(n_components=best_k, random_state=42)
    lda_best.fit(X_cluster)
    
    # ÊèêÂèñÊØè‰∏™‰∏ªÈ¢òÁöÑÂÖ≥ÈîÆËØç
    feature_names = vectorizer.get_feature_names_out()
    
    # ÊèêÂèñÊØè‰∏™‰∏ªÈ¢òÁöÑÂâç 20 ‰∏™ËØçÂèäÂÖ∂ÂæóÂàÜÔºåÂπ∂ËøõË°åÂèØËßÜÂåñ
    top_words = 20  # ÊØè‰∏™‰∏ªÈ¢òÊòæÁ§∫ÁöÑÂâç 20 ‰∏™ËØç
    
    # ÂàõÂª∫‰∏ÄË°å‰∫î‰∏™Â∞èÂõæÁöÑÂ∏ÉÂ±Ä
    fig, axes = plt.subplots(nrows=1, ncols=5, figsize=(20, 5))
    
    for topic_idx, topic in enumerate(lda_best.components_):
        if topic_idx < 5:  # Âè™ÁªòÂà∂Ââç‰∫î‰∏™‰∏ªÈ¢ò
            top_indices = topic.argsort()[-top_words:][::-1]
            top_features = [feature_names[i] for i in top_indices]
            top_scores = [topic[i] for i in top_indices]
    
            sns.barplot(x=top_scores, y=top_features, ax=axes[topic_idx])
            axes[topic_idx].set_title(f"Topic {topic_idx}")
            axes[topic_idx].set_xlabel("Score")
            axes[topic_idx].set_ylabel("Word")
    
    plt.tight_layout()
    plt.show()
    
    
    # Á°ÆÂÆöÊúÄÂº∫ÁöÑ‰∏ªÈ¢ò
    topic_distributions = lda_best.transform(X_cluster)
    dominant_topics = np.argmax(topic_distributions, axis=1)
    strongest_topic = np.bincount(dominant_topics).argmax()
    print(f"\nStrongest Topic: {strongest_topic}")
```

```python
from collections import Counter
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

# ‰∏∫ÊØè‰∏™ sentiment Ê†áÁ≠æÁîüÊàêËØçÈ¢ëÁªüËÆ°
sentiments = df['sentiment'].unique()
word_freq_by_sentiment = {}

# ËÆæÁΩÆ TF-IDF ÂêëÈáèÂô®
vectorizer = TfidfVectorizer(stop_words='english')

for sentiment in sentiments:
    # ÊèêÂèñËØ• sentiment ÁöÑÊâÄÊúâÊñáÊú¨
    sentiment_text = df[df['sentiment'] == sentiment]['text']

    # ËÆ°ÁÆó TF-IDF
    X = vectorizer.fit_transform(sentiment_text)
    feature_names = vectorizer.get_feature_names_out()
    tfidf_scores = np.sum(X, axis=0).A1
    
    # ÂàõÂª∫ËØçÈ¢ëÂ≠óÂÖ∏
    tfidf_dict = dict(zip(feature_names, tfidf_scores))

    # Â∞ÜÈ´ò TF-IDF ÂàÜÊï∞ÁöÑËØçÊ±áËßÜ‰∏∫ÂÖ≥ÈîÆËØçÔºåËÆ°ÁÆóÂÖ∂ËØçÈ¢ë
    word_freq = Counter({word: int(tfidf_scores[idx]) for word, idx in vectorizer.vocabulary_.items()})

    # Â≠òÂÇ®ËØçÈ¢ë
    word_freq_by_sentiment[sentiment] = word_freq
```
