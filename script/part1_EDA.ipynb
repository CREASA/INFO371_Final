```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("tweets.csv")
df
```

```python
# count the number of missing value in each column
nan_count = df.isna().sum()
print("Missing values in each column in the orignial dataset:")
nan_count
```

```python
df_new = df.fillna("NaN")
df_new
```

```python
# check after filling nan
nan_count_new = df_new.isna().sum()
print("Missing values in each column after filling NaN:")
nan_count_new
```

```python
num_rows, num_columns = df_new.shape

print("Number of rows:", num_rows)
print("Number of columns:", num_columns)
```

---
### Top users who post the most tweets

```python
top_users = df_new['user_name'].value_counts().head(10)
print("(2) Top 10 users with the most comments:")
print(top_users)
```

```python
import matplotlib.pyplot as plt
df_top_users = top_users.reset_index()
df_top_users.columns = ['User Name', 'Comments Count']

df_top_users_sorted = df_top_users.sort_values('Comments Count', ascending=True)

plt.figure(figsize=(10, 8))
plt.barh(df_top_users_sorted['User Name'], df_top_users_sorted['Comments Count'], color='skyblue')
plt.xlabel('Comments Count')
plt.ylabel('User Name')
plt.title('Top 10 Users with the Most Comments')
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.show()
```

---
### The change of tweets posted by day

```python
df_new['date'] = pd.to_datetime(df_new['date'], format='%Y-%m-%d %H:%M:%S%z', errors='coerce')
df_new = df_new.dropna(subset=['date'])
df_new['date'] = df_new['date'].dt.date
```

```python
import matplotlib.pyplot as plt

tweet_counts = df_new.groupby('date').size()

# Plotting
plt.figure(figsize=(10, 6))
tweet_counts.plot(kind='bar')
plt.title('Tweet Counts by Day')
plt.xlabel('Date')
plt.ylabel('Number of Tweets')
plt.xticks(rotation=45) 
plt.show()
```

We could see that the number of tweets is the highest on 2022-12-06, and this number decreased in general in the following days.

### Count unique descriptions

```python
unique_descriptions = df_new['user_description'].nunique()
print("Number of unique user descriptions:", unique_descriptions)
```

```python
df_description = df_new.copy()
df_description = df_new[df_new['user_description'] != "NaN"]
```

```python
top_des = df_description['user_description'].value_counts().head(10)
print("(2) Top 10 descriptions:")
print(top_des)
```

```python
df_top_des = top_des.reset_index()
df_top_des.columns = ['user_description', 'Comments Count']
df_top_des_sorted = df_top_des.sort_values('Comments Count', ascending=True)
```

```python
plt.rcParams['font.family'] = 'DejaVu Sans'

plt.figure(figsize=(12, 10))

plt.subplots_adjust(left=0.3) 

plt.barh(df_top_des_sorted['user_description'], df_top_des_sorted['Comments Count'], color='skyblue', height=0.8)

plt.xlabel('Comments Count')
plt.ylabel('User Description')
plt.title('Top 10 User Descriptions with the Most Comments')

plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.show()
```

---
### User followers intervals

```python
df_follower = df_new.copy()
df_follower['user_followers'] = pd.to_numeric(df_new['user_followers'], errors='coerce')
bins = [0, 100, 1000, 10000, 100000, 1000000, np.inf]
followers_intervals = pd.cut(df_follower['user_followers'], bins=bins)
followers_interval_counts = followers_intervals.value_counts()

print("User followers intervals:")
print(followers_interval_counts)
```

```python
plt.figure(figsize=(10, 8))
plt.bar(followers_interval_counts.index.astype(str), followers_interval_counts.values, color='skyblue')
plt.xlabel('Followers Intervals')
plt.ylabel('Number of Followers')
plt.title('User Followers Distribution')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

Most users has a the number of followers between 100 and 1000, while only a small amount of users have a number of followers larger than 100000. 

---
### User friends intervals

```python
df_friends = df_new.copy()
df_friends['user_friends'] = pd.to_numeric(df_new['user_friends'], errors='coerce')
bins = [0, 100, 1000, 10000, 100000, 1000000, np.inf]
friends_intervals = pd.cut(df_friends['user_friends'], bins=bins)
friends_interval_counts = friends_intervals.value_counts()

print("User user_friends intervals:")
print(friends_interval_counts)
```

```python
plt.figure(figsize=(10, 8))
plt.bar(friends_interval_counts.index.astype(str), friends_interval_counts.values, color='skyblue')
plt.xlabel('Friends Intervals')
plt.ylabel('Number of Friends')
plt.title('User Friends Distribution')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

As similar pattern shown above, most of the users have the number of friends between 100 to 1000, but only a small amount of users has number of friends larger than 10000.
